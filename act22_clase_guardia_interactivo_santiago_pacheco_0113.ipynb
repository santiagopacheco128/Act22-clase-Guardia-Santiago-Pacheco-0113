## clase: Guardia


```python
class Guardia:
    def __init__(self, nombre, telefono, edad, honestidad, comunicacion, disciplina):
        # Atributos de identidad
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

        # Atributos de perfil
        self.honestidad = honestidad
        self.comunicacion = comunicacion
        self.disciplina = disciplina

    # Funciones (Métodos)
    def vigila(self):
        print(f"\n[{self.nombre}]: Está realizando un recorrido de vigilancia. Nivel de disciplina: {self.disciplina}/10.")

    def cuida(self, lugar):
        print(f"[{self.nombre}]: Se encuentra custodiando el acceso a {lugar}. Reporte: Sin novedades.")

    def reportes(self, mensaje):
        print(f"[{self.nombre}] CENTRAL: '{mensaje}' (Enviado vía radio).")

# --- Solicitud de datos al usuario ---

print("--- REGISTRO DE NUEVO GUARDIA ---")
nom = input("Nombre del oficial: ")
tel = input("Teléfono: ")
ed  = input("Edad: ")
hons = input("Nivel de Honestidad (1-10): ")
com  = input("Nivel de Comunicación (1-10): ")
dis  = input("Nivel de Disciplina (1-10): ")

# Creación del objeto con los datos ingresados
oficial_usuario = Guardia(nom, tel, ed, hons, com, dis)

# --- Ejecución de resultados ---

print("\n" + "="*35)
print("       REPORTE DE ACTIVIDAD")
print("="*35)

oficial_usuario.vigila()
oficial_usuario.cuida("Entrada Principal")
msg = input(f"\nEscriba el reporte de radio para {nom}: ")
oficial_usuario.reportes(msg)

print("\n" + "-"*35)
print("Programa terminado")
print("Programa realizado por Santiago Pacheco NL = 43")
```
